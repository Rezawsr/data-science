# SEABORN
> Seaborn adalah library data visualzation python yang berbasis matplotlib dan terintegrasi dengan Pandas. seaborn memiliki tampilan visualisasi yang beragam dan lebih menarik
dibandingkan matplotlib

>seaborn merupakan bagian komuntias pydata yang merupakan komunitas open source data tools untuke developers dan user pythone

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Membuat Basic Plot

```python
#membuat list baru
list1 = [1, 2, 4, 7, 3]
```

```python
#membuat plot
sns.lineplot(data = list1, color = 'red')# isiplot
plt.title("judul")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.show()
warnings.filterwarnings("ignore")
```

> salah satu kelebihan Seaborn adalah Styling plot nya. Seaborn memiliki 5 jenis style yang dapat kita coba dalam melakukan visualisasi agar lebih menarik yaitu : **darkgrid, whitegrid, dark, white, dan ticks** dengan cara gunakan fungsi **set_theme()**

```python
#contoh styling darkgrid
plt.figure(figsize=(5,4))
sns.lineplot(data=list1, color = 'purple')
sns.set_style('whitegrid')
plt.title("ini judul")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.lineplot(data=list1, color = 'purple')
sns.set_style('darkgrid')
plt.title("ini judul")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
```

```python
sns.color_palette()
```

```python
#load beberarapa contoh dataset
tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
```

# DISTRIBUTION PLOT
> Seaborn memiliki beberapa plot yang dapat kita gunakan untuk memvisualisasikan distribution data yaitu **histplot, kdeplot dan displot.**

```python
# contoh hishplot
sns.histplot(titanic['age'])
plt.show()
```

```python
#contoh kdeplot
sns.kdeplot(titanic["age"])
plt.show()
```

```python
#contoh displot
sns.displot(titanic['age'], color='purple')
plt.show()
```

```python
#menggunakan subsplot
plt.figure(figsize=(5,5))
sns.displot(data=titanic, x='age',col='alive', row='sex',kde=True)
plt.suptitle("judul", fontsize = 20)
plt.tight_layout(pad=2)
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", col='smoker', kde="true")
plt.show()
```

# Scatter Plot

> Seaborn memiliki beberapa plot yang dapat digunakan untuk memvisualisasi sebaran data yaitu `scatterplot, regplot` dan `implot`.
> plot ini memeiliki pallete, beberapa diantaranya adalah;

> ![](https://i.stack.imgur.com/am6gN.png)

```python
# contoh scatterplot
plt.figure(figsize=(8,5))
sns.scatterplot(data=tips, x ='total_bill', y='tip',
                hue = "day", palette = "Set1", style='smoker')
plt.title("judul")
plt.show()
```

```python
#contoh regplot
sns.regplot(data=tips, x = 'total_bill', y = 'tip')
plt.show()
```

```python
#contoh implot
sns.lmplot(x='total_bill', y='tip', markers=['o', 'x'],
           data=tips, hue='smoker', palette="Set1")
plt.show()
```

```python
#menggunakan subplot
sns.lmplot(data=tips, x='total_bill', y ='tip', 
           col='smoker',  row='sex')

plt.suptitle("judul", fontsize = 20)
plt.tight_layout(pad=2)
plt.show()
```

```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="alive")
plt.show()
```

# BARPLOT dan BOXPLOT
> Seaborn memiliki plot yang dapat kita gunakan untuk memvisualisasikanjumlah data yaitu, `'countplot, barplot` dan `catplot`

> Selain itu juga dapat memvisualisasikan quantile dan outlier pada dataset denga `boxplot`

```python
# contouh countplot
no = tips['smoker'].value_counts()[0]
yes = tips['smoker'].value_counts()[1]

plt.figure(figsize=(5,5))
sns.countplot(data=tips, x='smoker', palette='Set1')
plt.title(f"smoker yes: {yes}||smoker no {no}")
plt.show()
```

```python
tips['smoker'].value_counts()
```

```python
titanic.info()
```

```python
sns.countplot(data=titanic, x='sex', palette ='Set1')
plt.show()
```

```python
titanic['sex'].value_counts()
```

```python
#contoh countplot dengan hue
sns.countplot(data=titanic, x='embark_town', hue='alive', palette='Set2')
plt.show()
```

```python
# nilai countplot
titanic.groupby('embark_town')["alive"].value_counts()
```

```python
# contoh barplot
sns.barplot(data=titanic, x='embark_town', y='fare', hue='alive', palette="Set2")
plt.show()
```

```python
titanic.groupby(['embark_town', 'alive'])['fare'].mean()
```

```python
#contoh barplot denga estimator
sns.barplot(data=titanic, x='embark_town', y='fare', hue='alive', palette ="Set2", estimator=np.sum)
plt.show()
```

```python
titanic.groupby(['embark_town', 'alive'])['fare'].sum()
```

```python
# contoh boxplot
sns.boxplot(data=tips, x='day', y='total_bill', palette ='Set2')
plt.show()
```

# HEATMAP CORELATION

> Seaborn memiliki fungsi `heatmap` yang sangat berguna unutk memvisualisasikan korelasi antar variabel numerik

> Heatmap adalah visualisasi atau pemetaan dengan menampilkan data dengan representasi warnay yang berbeda-beda, semakin tinggi angka suatu warnanya akan semakin gelap, dan begitu sebaliknya

```python
titanic.info()
```

```python
titanic.corr()
```

```python
#contoh heatmap
plt.figure(figsize=(12,10))
sns.heatmap(titanic.corr(), annot=True)
plt.show()
```

> - Pertama, koefisien korelasi antara 0,81 hingga 1,00 berarti hubungan antara dua variabel sangat kuat.
> - Kedua, koefisien korelasi antara 0,61 hingga 0,80 berarti hubungan antara dua variabel kuat. 
> - Ketiga, koefisien korelasi antara 0,41 hingga 0,60 berarti hubungan antara dua variabel sedang.

# JOINT PLOT DAN PAIR PLOT
