#### **Terdapat 3 jenis missing value:**
---
1. **MCAR (Missing Completely at Random)** atau data yang hilang tetapi tidak terkait dengan suatu variable, sehingga untuk kebanyakan kasus hal ini terjadi karena error, bisa human error atau corrupt pada data. **Cara handling nya adalah dengan drop, walaupun tetap disarankan diisi**. (hilangnya < 3%, sehingga tidak masalah untuk di drop)

2. **MAR (Missing at Random)** data yang hilang kemungkinan besar terkait dengan variablenya, atau biasanya dalam proses sampling / pengumpulan data, responden sengaja tidak menjawab pertanyaan untuk variable ini, entah mungkin karena memang privasi atau hat lain, sehingga nilainya menjadi kosong. **Cara handling nya adalah dengan missing values imputation**. (hilangnya > 10%, sehingga tidak boleh di drop karena akan berpengaruh, biasanya memakai mean (ada proxy, contohnya ada data gaji pekerja di bidang IT dan lama kerja 5 tahun yang hilang, maka kita harus memakai rata-rata dari data pekerja IT yang lama kerjanya 5 tahun juga))

3. **MNAR (Missing Not at Random)** merupakan data yang hilang memang bukan karena error atau terkait dengan variable tertentu seperti pada MCAR atau MAR. Terjadinya missing data pada suatu variabel berkaitan dengan variabel itu sendiri, sehingga ini tidak bisa diprediksi dari variabel lain pada suatu dataset. Data yang hilang pada hal ini memang bisa jadi adalah nilai dari suatu data (value = Null). **Cara handling nya adalah dengan mengubah ke dalam isi yang dimaksud, dalam case tertentu Null ini dapat diisi dengan 0 atau unknown**. (isi data tersebut memang null, bukan tidak terisi)
