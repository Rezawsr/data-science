## 1. Membuat Basic Plot

> Melakukan plot grafik yang menghubungkan dua poin. Kita dapat gunakan fungsi `plot()`.

> Ini hanya membutuhkan suatu array atau list untuk melakukan plot 2D (X vs Y).

```python
# contoh plot sederhana
list1 = [1, 2, 3, 7, 3]
```

```python
# buat plot
plt.plot(list1)
plt.show()
```

> Kita dapat gunakan `figure()` dan `figsize()` untuk merubah size plot, `title()`, `xlabel()`, dan `ylabel()` untuk menambahkan keterangan judul dan setiap axisnya.

```python
# contoh title dan Label
plt.figure(figsize=(7, 5))  # merubah size
plt.plot(list1)
plt.title("Ini Judul Plot")
plt.xlabel("Ini X axis")
plt.ylabel("Ini Y axis")
plt.show()
```

```python
# menambahkan list
list2 = [2, 3, 5, 3, 4]
```

```python
# contoh title dan Label
plt.figure(figsize=(7, 5))  # merubah size
plt.plot(list1)
plt.plot(list2)
plt.title("Ini Judul Plot")
plt.xlabel("Ini X axis")
plt.ylabel("Ini Y axis")
plt.show()
```

> Kita dapat menggunakan styling tertentu untuk menambah kesan estetika dalam plot, walaupun tidak terlalu disarankan. Dan melakukan saving / menyimpan plot yang sudah kita buat dengan `savefig()`.

> Lebih lanjut: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

```python
# contoh masukkan list baru ke plot t styling dan save
plt.figure(figsize=(7, 5))
plt.plot(list1, linewidth=2, linestyle='--', marker='*', color="red")
plt.plot(list2, linewidth=1, linestyle='-', marker='.', color="green")
plt.title("Ini Judul Plot")
plt.xlabel("Ini X axis")
plt.ylabel("Ini Y axis")
plt.savefig("contoh_plot.png")  # save plot
plt.show()
```

# 2. BEBEERAPA JENIS PLOT

> Selain kaena mudah dipelajari matploib memiliki banyak contoh grafik yang bisa kita gunakan untuk memvisualisasikan data agar menjadi lebih menarik, Antara lain seperti 

> 1. Line plot - jenis grafik untuk menampilkan informasi dengan menggunakan banyak titik yang saling terhubung dan membentuk garis lurus atau lengkung biasanya, digunakan untuk visualisasi timeseries

> 2. Bar plot - merupakan jenis grafik yang paling sering digunakan untuk mempresentasikan data numerik dan kategorik dalam bentuk bar.

> 3. scatter plot - merupakan plot yang dapat kita gunakan untuk mencari korelasi antara dua tabel (x dari y),

> 4. Stacked plot - untuk plot data linier dalam urutan vertikal, stacking pada setiap plotnya. biasanya digunakan untuk menggunakan plot kumulatif

> 5. Boxplot - Membuat bentuk visualisasi data yang digunakan untuk memeriksa keberadaan outlier melalui lima dimensi utam yaitu nilai min, max, ,Q1,Q2,Q3.

# Line Plot

> Data https://www.cryptodatadownload/com/data/binance/

```python
#load data
stock = pd.read_csv("binance.csv", delimiter = ",")
```

```python
# cek head
stock.head()
```

```python
stock.info()
```

```python
# file row columns
stock2 = stock[["date", "close"]]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# ubah data ke datetime
stock2 ["date"] = pd.to_datetime(stock2["date"])
```

```python
stock2.info()
```

```python
#sort date
stock2 = stock2.sort_values(by="date").reset_index(drop=True)
```

```python
# ganti tipe data
stock2["close"] = stock2["close"].astype(float)
```

```python
plt.figure(figsize=(8,5))
plt.plot(stock2["date"][:100], stock2["close"][:100])
plt.title("Algo Bitcoin Price")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.xticks(rotation=45)
plt.show()
```

# Bar Plot

> Terdapat dua cara untuk menggunakan barplot, dengan fungsi bar() dan barh()

> data https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv

```python
iris = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

```python
iris.head()
```

```python
iris.info()
```

```python
# rubah nama columns
cols = []
for i in iris.columns.tolist():
    cols.append(i.replace(".", "_"))

iris.columns = cols
```

```python
# cek perubahan
iris.info()
```

```python
plt.figure(figsize=(8,5))
plt.bar(iris["variety"], iris["sepal_length"])
plt.title("Iris")
plt.xlabel("Variety")
plt.ylabel("Sepal Length")
plt.xticks(rotation=20)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.barh(iris["variety"], iris["sepal_length"])
plt.title("Iris")
plt.xlabel("Variety")
plt.ylabel("Sepal Length")
plt.xticks(rotation=20)
plt.show()
```

# Scatter Plot
> Untuk menggambar plot scatter, kita dapat gunakan fungsi scatter().

```python
s = []
for i in iris["variety"]:
    if i == "setosa":
        s.append(0)
    elif i == "versicolor":
        s.append(1)
    else :
        s.append(2)
```

```python
#masukan hasil ke kolom baru
iris["variety_enc"] = s
```

```python
iris.head()
```

```python
plt.scatter(iris["petal_length"], iris["petal_width"], c=iris["variety_enc"], alpha=0.6)
plt.title("Iris Variety")
plt.xlabel("Petal Widht")
plt.ylabel("Petal Length")
plt.xticks(rotation=20)
plt.show()
```

# BOXPLOT
> Boxplot biasa digunakan untuk melihat distribusi data dan outlier, kita bisa gunakan fungsi boxplot().

> Data https://raw.githubusercontent.com/mlflow/mlflow-example/master/wine-quality.csv

```python
wine = pd.read_csv("https://raw.githubusercontent.com/mlflow/mlflow-example/master/wine-quality.csv")
```

```python
wine
```

```python
#membuat boxplot
plt.boxplot([wine["fixed acidity"],wine["alcohol"], wine['quality']])
plt.show()
```

```python
#vertikal boxplot
plt.boxplot([wine["fixed acidity"],wine["alcohol"], wine['quality']], vert=False)
plt.show()
```

# SUBPLOT

```python
# contoh subplot

fig = plt.figure()

ax1 = plt.subplot(221) #nrows, ncols, pos
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.title.set_text("judul 1")# isi dari axis
ax2.title.set_text("judul 2")
ax3.title.set_text("judul 3")
ax4.title.set_text("judul 4")

ax1.plot([1,2,2])# judul dari axis
ax2.plot([2,1,7])
ax3.plot([4,12,8])
ax4.plot([2,5,9])

plt.tight_layout() # untuk jarak antar subplot
plt.show();
```

```python
# buat mean dan std untuk data iris
iris_mean = iris.mean()
iris_std = iris.std()
```

```python
iris_mean[:4]
```

```python
iris_std[:4]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 8))
axs[0, 0].hist(iris["petal_width"])
axs[0, 0].set(title = "Plot Histogram", xlabel= "Petal Bins", ylabel= "Petal Width")

axs[0, 1].plot(iris["sepal_length"], iris["sepal_width"])
axs[0, 1].set(title = "Plot Line", xlabel= "Sepal Length", ylabel= "Sepal Width")

axs[1, 0].scatter(iris["petal_length"], iris["petal_width"])
axs[1, 0].set(title = "Plot Scatter", xlabel= "Petal Length", ylabel= "Petal Width")

axs[1, 1].bar(["sepal_length","sepal_width", "petal_length","petal_width"], iris_mean[:4], yerr= iris_std[:4])
axs[1, 1].set(title = "Plot Bar", xlabel= "Category", ylabel= "Category Mean")


plt.suptitle("judul plot")
plt.tight_layout(pad=2, w_pad=1.0, h_pad=1.0)
plt.xticks(rotation= 45)
plt.show()
```
